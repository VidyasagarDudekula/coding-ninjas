```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
train_data = pd.read_csv('/Users/dvidyasagar/Coding/ML/coding ninjas/data/train_diabetes.csv')
train_data.head()
```

```python
train_data.drop('Y', axis = 1)
```

```python
X_train = np.array(train_data.drop('Y', axis = 1))
Y_train = np.array(train_data['Y'])
print(X_train.shape, Y_train.shape)
```

```python
algo = LinearRegression()
```

```python
algo.fit(X_train, Y_train)
```

```python
algo.score(X_train, Y_train)
```

```python
test_data = np.genfromtxt('/Users/dvidyasagar/Coding/ML/coding ninjas/data/test_diabetes.csv', delimiter=',')
print(len(test_data))
X_test = np.array(test_data)
Y_predict = algo.predict(X_test)
print(Y_predict.shape, X_test.shape)
```
