```python
#we can add a line width.
x = np.arange(0,8,0.1)
y = cube(x)

plt.plot(x,y, color = 'blue', linewidth = 0.5)
plt.show()
```

```python
#we can add a line width.
x = np.arange(0,8,0.1)
y = cube(x)

plt.plot(x,y, color = 'blue', linewidth = 0.5)
plt.xlabel('X')
plt.ylabel('X^3')
plt.title("Matplotlib Demo")
plt.show()
```

```python
#we can add a line width.
def square(x):
    return x**2

x = np.arange(0,15,0.1)
y = cube(x)
y1 = square(x)
plt.plot(x,y, color = 'black', linewidth = 1)
plt.plot(x,y1, color = 'red', linewidth = 1)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend() #defines each line in graph
plt.grid()
plt.title("Matplotlib Demo")
plt.show()
```

```python
#lets discuss the pie graphs.
#the mininum we have to give to the pie graph is sizes.
#lets say i have four different parts in the pie graph, so lets give them the sizes.
sizes = [3,4,6,2]
plt.pie(sizes)
plt.show()
```

```python
sizes = [3, 4, 6, 2]
colors=['green', 'purple', 'red', 'pink']
labels = ['A', 'B', 'C', 'E']
plt.title('Split among the classes')
plt.pie(sizes, colors=colors, labels=labels)
plt.show()
```

```python
sizes = [3, 4, 6, 2]
colors=['green', 'purple', 'red', 'pink']
labels = ['A', 'B', 'C', 'E']
plt.title('Split among the classes')
plt.pie(sizes, colors=colors, labels=labels, autopct="%.2f%%")
plt.show()
```

```python
#if you want to hight light a perticular data. like exploring that data.
#we take a pie --> explode
sizes = [3, 4, 6, 2]
colors=['green', 'purple', 'red', 'pink']
labels = ['A', 'B', 'C', 'E']
explode = [0.2, 0, 0, 0]
plt.title('Split among the classes')
plt.pie(sizes, colors=colors, labels=labels, explode=explode, autopct="%.2f%%")
plt.show()
```

```python
#lets make the colock clock wize direction.
sizes = [3, 4, 6, 2]
colors=['green', 'purple', 'red', 'pink']
labels = ['A', 'B', 'C', 'E']
explode = [0.2, 0, 0, 0]
plt.title('Split among the classes')
plt.pie(sizes, colors=colors, labels=labels, explode=explode, autopct="%.2f%%", counterclock=False)
plt.show()
```

```python
#Also the it is starting from the zeroth andle, so lets make it start at 90 degrees.
sizes = [3, 4, 6, 2]
colors=['green', 'purple', 'red', 'pink']
labels = ['A', 'B', 'C', 'E']
explode = [0.2, 0, 0, 0]
plt.title('Split among the classes')
plt.pie(sizes, colors=colors, labels=labels, explode=explode, autopct="%.2f%%", counterclock=False, startangle=90)
plt.show()
```
